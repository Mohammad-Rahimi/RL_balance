```python
import numpy as np
import skvideo.io
from base64 import b64encode
from IPython.display import HTML
from myosuite.utils.quat_math import quat2euler
from myosuite.utils import gym
import myosuite
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
import torch

def show_video(video_path, video_width=500):
    video_file = open(video_path, "r+b").read()
    video_url = f"data:video/mp4;base64,{b64encode(video_file).decode()}"
    return HTML(f"""<video autoplay width={video_width} controls><source src="{video_url}"></video>""")

# Create and wrap the environment
env = gym.make('myoLegWalk-v0', reset_type='random')

# Define the PPO algorithm
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = PPO(
    'MlpPolicy',
    env,
    verbose=1,
    tensorboard_log="./ppo_myo_tensorboard/",
    device=device
)

# Print PPO parameters
print("PPO Algorithm Parameters:")
print(model.policy)

# Print neural network details
print("\nNeural Network Details:")
print(model.policy.mlp_extractor.policy_net)
print(model.policy.mlp_extractor.value_net)

# Define a custom callback to access the reward function
class RewardCallback(BaseCallback):
    def __init__(self, verbose=0):
        super(RewardCallback, self).__init__(verbose)

    def _on_step(self) -> bool:
        rewards = self.locals['rewards']
        print(f"Rewards: {rewards}")
        return True

# Train the model
timesteps = 1_000_000
reward_callback = RewardCallback()
model.learn(total_timesteps=timesteps, callback=reward_callback)

# Save the model
model.save("ppo_myo_model3")
```
